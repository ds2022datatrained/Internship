```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
dset=driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
dset.click()
```

```python
UCI_URLS=[]
start=0
end=2
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        UCI_URLS.append(i.get_attribute("href"))
    n_button=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
    n_button.click()
    time.sleep(3)
```

```python
print(len(UCI_URLS))
```

```python
UCI_URLS
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
No_Of_Instances=[]
No_Of_Attribute=[]
Year=[]
```

```python
for url in UCI_URLS:
    driver.get(url)
    time.sleep(3)
    
    try:
        dataset_name=driver.find_elements(By.XPATH,'//div[@class="flex flex-col"]/div[1]/div[2]/div/h1')
        for i in dataset_name:
            d_name=i.text
            Dataset_Name.append(d_name)
    except NoSuchElementException:
        Dataset_Name.append('NA')
        
    try:
        data_type=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        for i in data_type:
            d_type=i.text
            Data_Type.append(d_type)
    except NoSuchElementException:
        Data_Type.append('NA')
        
    try:
        task=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        for i in task:
            tsk=i.text
            Task.append(tsk)
    except NoSuchElementException:
        Task.append('NA')
        
    try:
        attribute_type=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        for i in attribute_type:
            at=i.text
            Attribute_Type.append(at)
    except NoSuchElementException:
        Attribute_Type.append('NA')
        
    try:
        no_of_instances=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        for i in no_of_instances:
            noi=i.text
            No_Of_Instances.append(noi)
    except NoSuchElementException:
        No_Of_Instances.append('NA')
        
    try:
        no_of_attribute=driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        for i in no_of_attribute:
            noa=i.text
            No_Of_Attribute.append(noa)
    except NoSuchElementException:
        No_Of_Attribute.append('NA')
        
    try:
        year=driver.find_elements(By.XPATH,'//div[@class="relative flex flex w-full items-center gap-4 bg-primary p-2"]/div[2]/h2')
        for i in year:
            ye=i.text
            Year.append(ye)
    except NoSuchElementException:
        Year.append('NA')
```

```python
print(len(Dataset_Name),len(Data_Type),len(Task),len(Attribute_Type),len(No_Of_Instances),len(No_Of_Attribute),len(Year))
```

```python
Year.insert(4,'NA')
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://www.naukri.com/hr-recruiters-consultants")
```

```python
JOB_URLS=[]
start=0
end=1
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
    for i in url:
        JOB_URLS.append(i.get_attribute("href"))
    n_button=driver.find_element(By.XPATH,'//div[@class="pagination lastCompMark"]/a[2]/span')
    n_button.click()
    time.sleep(3)
```

```python
JOB_URLS
```

```python
print(len(JOB_URLS))
```

```python
Designation=[]
Company=[]
Experience=[]
CTC=[]
Location=[]
Posted_On=[]
Openings=[]
```

```python
for url in JOB_URLS:
    driver.get(url)
    time.sleep(3)
    
    try:
        designation=driver.find_elements(By.XPATH,'//div[@class="leftSec"]/section[1]/div[1]/div[1]/header/h1')
        for i in designation:
            des=i.text
            Designation.append(des)
    except NoSuchElementException:
        Designation.append('NA')
        
    try:
        company=driver.find_elements(By.XPATH,'//div[@class="leftSec"]/section[1]/div[1]/div[1]/div[1]')
        for i in company:
            com=i.text
            Company.append(com)
    except NoSuchElementException:
        Company.append('NA')
        
    try:
        experience=driver.find_elements(By.XPATH,'//div[@class="left"]/div[1]/div[1]/span')
        for i in experience:
            exp=i.text
            Experience.append(exp)
    except NoSuchElementException:
        Experience.append('NA')
        
    try:
        ctc=driver.find_elements(By.XPATH,'//div[@class="left"]/div[1]/div[2]/span')
        for i in ctc:
            ct=i.text
            CTC.append(ct)
    except NoSuchElementException:
        CTC.append('NA')
        
    try:
        location=driver.find_elements(By.XPATH,'//div[@class="left"]/div[2]/span')
        for i in location:
            loc=i.text
            Location.append(loc)
    except NoSuchElementException:
        Location.append('NA')
     
    try:
        posted_on=driver.find_elements(By.XPATH,'//div[@class="bottom"]/div/span[1]/span')
        for i in posted_on:
            pos=i.text
            Posted_On.append(pos)
    except NoSuchElementException:
        Posted_On.append('NA')
        
    try:
        openings=driver.find_elements(By.XPATH,'//div[@class="bottom"]/div/span[2]/span')
        for i in openings:
            opn=i.text
            Openings.append(opn)
    except NoSuchElementException:
        Openings.append('NA')
```

```python
print(len(Designation),len(Company),len(Experience),len(CTC),len(Location),len(Posted_On),len(Openings))
```

```python
hr_recruiters_consultants=pd.DataFrame({'Designation':Designation,'Company':Company,'Experience':Experience,'CTC':CTC,'Location':Location,'Posted On':Posted_On,'Openings':Openings})
```
